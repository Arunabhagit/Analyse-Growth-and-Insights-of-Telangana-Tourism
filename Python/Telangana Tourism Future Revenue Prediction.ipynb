```python
data = {
    'year': [2016, 2017, 2018, 2019],
    'domestic': [95160830, 85266596, 92878329, 83035894],
    'foreign': [166570, 251846, 318154, 323326]
}
```

```python
df = pd.DataFrame(data)
df['total_visitors'] = df['domestic'] + df['foreign']
```

```python
df['domestic_proportion'] = df['domestic'] / df['total_visitors']
df['foreign_proportion'] = df['foreign'] / df['total_visitors']
```

```python
avg_domestic_proportion = df['domestic_proportion'].mean()
avg_foreign_proportion = df['foreign_proportion'].mean()

# Total Visitor Predicted data for 2025-2030
future_visitors = {
    'year': [2025, 2026, 2027, 2028, 2029, 2030],
    'total_visitors': [91040845, 93277911, 95774152, 98394704, 100902943, 102961753]
}
```

```python
future_df = pd.DataFrame(future_visitors)
```

```python
# Predict the number of Domestic and Foreign visitors for each year (2025-2030)
future_df['domestic_visitors'] = future_df['total_visitors'] * avg_domestic_proportion
future_df['foreign_visitors'] = future_df['total_visitors'] * avg_foreign_proportion
```

```python
domestic_price = 1200 #Average price for domestic visitors
foreign_price = 5600  #Average price for Foreign visitors

future_df['domestic_revenue'] = future_df['domestic_visitors'] * domestic_price
future_df['foreign_revenue'] = future_df['foreign_visitors'] * foreign_price
```

```python
future_df['total_revenue'] = future_df['domestic_revenue'] + future_df['foreign_revenue']
```

```python
def format_million(value):
    return f"{value / 1e6:.2f}M"
```

```python
future_df['domestic_visitors'] = future_df['domestic_visitors'].apply(format_million)
future_df['foreign_visitors'] = future_df['foreign_visitors'].apply(format_million)
future_df['domestic_revenue'] = future_df['domestic_revenue'].apply(format_million)
future_df['foreign_revenue'] = future_df['foreign_revenue'].apply(format_million)
future_df['total_revenue'] = future_df['total_revenue'].apply(format_million)

# Display the results
print(future_df[['year', 'domestic_visitors', 'foreign_visitors', 'domestic_revenue', 'foreign_revenue', 'total_revenue']])
```
